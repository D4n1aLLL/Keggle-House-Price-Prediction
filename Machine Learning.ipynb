# Training for kaggle submission.

### Using whole train.csv for training and predict values based on data in test.csv 

```python
abc = GradientBoostingRegressor(n_estimators=5000,learning_rate=0.01)
abc.fit(X, Y)
abc_predict = abc.predict(test)
```

### Submission file needs Id and SalePrice, so I am fetching Id from samples and concatenating it with predictions

```python
prediction_df = pd.DataFrame(abc_predict)
samples = pd.read_csv("sample_submission.csv")
final = pd.concat([samples['Id'],prediction_df],axis=1)
final.columns = ["Id","SalePrice"]
```

```python
final.to_csv("abc.csv",index=False)
```

## Training and testing in train.csv only

### test.csv doesnt have SalePrice columns which makes it is impossible to measure score using both datasets, therefore using train.csv only for both traininig and testing

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
abc = GradientBoostingRegressor(n_estimators=100000,learning_rate=0.01,verbose=1)
abc.fit(x_train, y_train)
abc_predict = abc.predict(x_test)
print("Model Score: " + str(abc.score(x_test,y_test)))
```

### Selecting random Prices from prediction column and actual values to see the difference

```python
import random
indexes = []
actual = y_test.iloc[:].values
for i in range(0,20):
    temp = random.randint(0,len(abc_predict)-1)
    if temp not in indexes:
        print("Predicted: "+str(abc_predict[temp])+" - Actual: "+str(actual[temp]))
        indexes.append(temp)
```

```python
import sklearn.metrics as m
```

```python
print("Mean Squared Error: " + str(m.mean_squared_error(y_test,abc_predict)))
print("Log Mean Squared Error: " + str(m.mean_squared_log_error(y_test,abc_predict)))
```
